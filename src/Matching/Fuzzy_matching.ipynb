```python
d=pd.DataFrame(dictionary[' description'])
d.columns = ['company_names']
```

```python
df=d.drop_duplicates()
df.info()
```

```python
fda['FDA_companies'] = fda['FDA_companies'].str.lower()
df['company_names'] = df['company_names'].str.lower()
df1=df[:1000]
```

```python
match_companies=[]
for i in df1['company_names']:
    match=process.extractOne(i, fda.FDA_companies)
    match_companies.append(i)
    match_companies.append(match)
```

```python
match_companies[:50]
```

```python
def match(Col1,Col2):
    overall=[]
    for n in Col1:
        result=[(fuzz.ratio(n, n2),n2)
                for n2 in Col2 if fuzz.ratio(n, n2)>50
               ]
        if len(result):
            result.sort()
            print('result {}'.format(result))
            print("Best M={}".format(result[-1][1]))
            overall.append(result[-1][1])
        else:
            overall.append(" ")
    return overall

print(match(fda.FDA_companies,df1.company_names))
```
