#### Creating a list of punctuation

```python
removeset = string.punctuation
removeset = removeset.replace("-","") #Don't remove dashes, why? common connector in company name
removeset = removeset.replace("&","") #Don't remove ampersand, why? common connector in company name
removeset = removeset.replace("_","") #Don't remove underscore, why? common connector in company name

print(removeset)
```

Using the list above to remove the punctuation

```python
#function that gets rid of unwanted punctuation
def removePunct(string):
    return re.sub('[!"#\'()*+,./:;<=>?@[\]^`{|}~]', '', string)
```

```python
dna_dict['cleaned_companies'] = dna_dict['cleaned_companies'].apply(removePunct)
```

```python
dna_dict.head()
```

#### Removing all single characters

```python
def  removeSingle(string):
    return re.sub(r'\s+[a-zA-Z]\s+', '', string)
```

```python
dna_dict['cleaned_companies'] = dna_dict['cleaned_companies'].apply(removeSingle)
```

```python
dna_dict.iloc[50:100]
```

#### Removing all numbers

```python
def removeNum(string):
    return re.sub(r'[0-9]','', string)
```

```python
dna_dict['cleaned_companies'] = dna_dict['cleaned_companies'].apply(removeNum)
```

```python
dna_dict.iloc[640:660]
```

#### Removing the content in the parentheses along with the parentheses
(Function uses regex to remove parentheses)

```python
def removeParenthesesContent(string):
    return re.sub(r'\([^)]*\)', '', string)
```

```python
dna_dict['cleaned_companies'] = dna_dict['cleaned_companies'].apply(removeParenthesesContent)
```

#### Removing single characters from the start

```python
def removeSingleChar(string):
    return re.sub(r'\^[a-zA-Z]\s+', ' ', string)
```

```python
dna_dict['cleaned_companies'] = dna_dict['cleaned_companies'].apply(removeSingleChar)
```

```python
dna_dict.iloc[900:950]
```

#### Substituting multiple spaces with single space

```python
def removeMultiSpace(string):
    return re.sub(r'\s+', ' ', string, flags=re.I)
```

```python
dna_dict['cleaned_companies'] = dna_dict['cleaned_companies'].apply(removeMultiSpace)
```

```python
dna_dict.iloc[50:70]
```

#### Converting to lowercase

```python
dna_dict['cleaned_companies'] = dna_dict['cleaned_companies'] .str.lower()
```

```python
dna_dict.iloc[200:220]
```

#### Removing prefixed 'b'

```python
def removePrefixb(string):
    return re.sub(r'^b\s+', '', string)
```

```python
dna_dict['cleaned_companies'] = dna_dict['cleaned_companies'].apply(removePrefixb)
```

Making dashes, ampersands and underscores into combined words

```python
def combWords1(string):
    return re.sub(r'\s-\s+', '-', string)
```

```python
dna_dict['cleaned_companies'] = dna_dict['cleaned_companies'].apply(combWords1)
```

```python
def combWords2(string):
    return re.sub(r'\s&\s+', '&', string)
```

```python
dna_dict['cleaned_companies'] = dna_dict['cleaned_companies'].apply(combWords2)
```

```python
def combWords3(string):
    return re.sub(r'\s_\s+', '_', string)
```

```python
dna_dict['cleaned_companies'] = dna_dict['cleaned_companies'].apply(combWords3)
```

#### Which word tokens appeared in the dataset most often

```python
#cat all the row entries into one long string
longString=dna_dict['Description'].str.cat(sep=' ')

#separate each "word" (space separated token) into a extremely long list
longStringSeparated=longString.split(' ')

#turn it into a dataframe
uniqueSubTokenFrame=pd.DataFrame(longStringSeparated)

#get the count on that column, this tells us the frequency of those unique tokens
columnUniqueCounts=uniqueSubTokenFrame.iloc[:,0].value_counts()

#convert that output to a proper table
tableUniqueCounts=columnUniqueCounts.reset_index()
tableUniqueCounts.rename(columns={0:"count","index":"token"},inplace=True)

print(tableUniqueCounts.shape)
print('number of unique string tokens in this dataset')

tableUniqueCounts.head(50)
```

#### Reading in the set of legal entities listed by OSS team

```python
legalEntities = pd.read_csv("https://raw.githubusercontent.com/DSPG-Young-Scholars-Program/dspg20oss/danBranch/ossPy/keyFiles/curatedLegalEntitesRaw.csv", quotechar = "'",header = None)
```

```python
legalEntities
```

```python
def eraseFromColumn(inputColumn,eraseList):
     """iteratively delete regex query matches from input list
    
    Keyword arguments:
    inputColumn -- a column from a pandas dataframe, this will be the set of
    target words/entries that deletions will be made from
    eraseList -- a column containing strings (regex expressions) which will be
    deleted from the inputColumn, in an iterative fashion
    """
    eraseList['changeNum']=0
    eraseList['changeIndexes']=''
    inputColumn=inputColumn.replace(regex=True, to_replace='\\\\',value='/')
    for index, row in eraseList.iterrows():
        curReplaceVal=row[0]
        currentRegexExpression=re.compile(curReplaceVal)
        CurrentBoolVec=inputColumn.str.contains(currentRegexExpression,na=False)
        eraseList['changeIndexes'].iloc[index]=[i for i, x in enumerate(CurrentBoolVec) if x]
        eraseList['changeNum'].iloc[index]=len(eraseList['changeIndexes'].iloc[index])
        inputColumn.replace(regex=True, to_replace=currentRegexExpression,value='', inplace=True)

    return inputColumn;
```

```python
dna_dict['cleaned_companies'] = eraseFromColumn(dna_dict['cleaned_companies'],legalEntities)
```

#### Finding common words to remove

```python
#cat all the row entries into one long string
longString=dna_dict['cleaned_companies'].str.cat(sep=' ')

#separate each "word" (space separated token) into a extremely long list
longStringSeparated=longString.split(' ')

#turn it into a dataframe
uniqueSubTokenFrame=pd.DataFrame(longStringSeparated)

#get the count on that column, this tells us the frequency of those unique tokens
columnUniqueCounts=uniqueSubTokenFrame.iloc[:,0].value_counts()

#convert that output to a proper table
tableUniqueCounts=columnUniqueCounts.reset_index()
tableUniqueCounts.rename(columns={0:"count","index":"token"},inplace=True)

print(tableUniqueCounts.shape)
print('number of unique string tokens in this dataset')

tableUniqueCounts.head(20)
```

```python
stopwords=['of', 'group', 'holdings', 'international', 'the', 'and', 'capital', 'health', 'medical', 'technologies', 'bank', 'universities', 'energy', 'partners', 'association', 'services', 'national', 'systems', 'american', 'phermaceuticals']
stopwords
```

```python
dna_dict['cleaned_companies']=dna_dict['cleaned_companies'].apply(lambda x: ' '.join([word for word in x.split() if word not in stopwords]))
```
